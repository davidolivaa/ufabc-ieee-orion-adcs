```python
#bibliotexas para o projeto
import control as ctl
import matplotlib.pyplot as plt
import numpy as np
```

```python
#variaveis de resistencia e capacitancia
R = 1.;
C = 1.;

#vetores que serao os polinomios para a funçao de transferencia
Tsim = 10 #vetor de tempo de simulaçao
numerador = [1.]
denominador = [R*C, 1.]
```

```python
#usa os vetores para criar a funçao de transferencia
H = ctl.tf(numerador, denominador)
print('FT malha aberta:', H)
```

```python
#calcula a resposta ao degrau
T, yout = ctl.step_response(H, Tsim) # retorna dois valores a partir dos parametros de funçao de transferencia e tempo
plt.plot(T, yout, 'b-')

#calcula um degrau unitario
T2 = np.linspace(-1., 10., 1000)
degrau = np.ones_like(T2)
degrau[T2<0] = 0
plt.plot(T2, degrau, 'r-')

#melhorando o grafico
plt.ylabel('V (V)')
plt.xlabel('Tempo (s)')
plt.legend(['resposta ao degrau', 'degrau unitario'])
plt.grid()
```
